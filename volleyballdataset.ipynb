```python
ln=['Silvija POPOVIC', 'Daria PILIPENKO', 'Michaela SPANKOVA', 'Leja JANEŽIČ', 'Joana RESENDE']
lr=['98', '9', '9', '9', '8']
mn=[ 'Kseniya PAVLENKO', 'Hristina VUCHKOVA', 'Ajla PARADŽIK', 'Lucia VARELA GOMEZ', 'Nadzeya STOLIAR']
mr=['99', '99', '98', '95', '86']
plt.plot(mn,mr,color='#32A6BF',linewidth=3)
plt.plot(ln,lr,color='#A0E0EF',linewidth=3)
plt.xlabel("PLAYERS")
plt.ylabel("RECEPTİON")
plt.title( "VOLLEYBALL DATA")
plt.legend(["MidPlayers","Liberos"])


plt.style.use('seaborn-v0_8-pastel')
plt.legend()
plt.grid()


plt.gcf().autofmt_xdate(rotation=45)  #hata duzeltme fonk
plt.show()
```

```python
plt.bar(ln,lr,color='#041B46',label="liberos")
plt.bar(mn,mr,color="#4B82A2",label="middle blockers")
plt.title("Playes' Reception Rate")
plt.xlabel("players")
plt.ylabel("reception")
plt.tight_layout()
plt.grid()
plt.gcf().autofmt_xdate(rotation=25)  #hata duzeltme fonk

plt.show()
```
